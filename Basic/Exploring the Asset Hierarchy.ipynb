```python
%pylab notebook
import pandas as pd
import os

from cognite.config import configure_session
from cognite.v05.assets import get_assets, get_asset_subtree
```

```python
# Replace 'CDP_key' with your environment varible
configure_session(os.environ.get('CDP_key'), 'akerbp')
```

```python
# First, let us find the Ivar Aasen platform
DTO = get_assets(name='IAA') 
# The DTO is the response object. To make it accesible in a way 
# we are used to, we convert it to a Pandas DataFrame
root_node = DTO.to_pandas()
print(root_node)
```

```python
# Next, let us list the subsystems at Ivar Aasen
IAA_id = root_node.id.iloc[0]
DTO = get_asset_subtree(asset_id = IAA_id, depth=1)
assets = DTO.to_pandas()
assets
```

```python
# We can continue to dive down into the Main Process Systems
Main_process_id = assets.id.iloc[4]
DTO = get_asset_subtree(asset_id = Main_process_id, depth=1)
DTO.to_pandas()
```

```python
# Let us see all assets under GAS TREATMENT
DTO = get_asset_subtree(asset_id = 6268750830585918, depth=10)
DTO.to_pandas()
```

```python
# Next, let us locate a scrubber on Ivar Aasen (https://en.wikipedia.org/wiki/Scrubber)
DTO = get_assets(description='Scrubber', path='[{}]'.format(IAA_id))
assets = DTO.to_pandas()
assets
```

```python
# Let us review the information available about the 1st suction scrubber
scrubber = assets.iloc[20]
scrubber.metadata
```

```python
# Next, let us extract the description of all nodes down to the scrubber
path = []
for aid in scrubber.path:
    ast = get_asset(asset_id=aid).to_pandas()
    path.append(ast.loc['description'].values[0])
print(path)
```

```python
# Are there any sub-assets in the scrubber?
DTO = get_asset_subtree(asset_id = scrubber.id)
sub_assets = DTO.to_pandas()
sub_assets
```
